```python
import cv2

# 이미지 읽기
image = cv2.imread('image.png')
```

```python
# 이미지 표시
cv2.imshow('Image', image)
cv2.waitKey(0)  # 키 입력이 있을 때까지 대기
cv2.destroyAllWindows()  # 열린 모든 창 닫기
```

```python
# 그레이스케일 변환 : RGB 컬러 이미지 -> 1채널 흑백 이미지
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Gray Image', gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 크기 조절
width, height = 512, 512
resized_image = cv2.resize(image, (width, height))

cv2.imshow('Resized image', resized_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 회전 (시계 방향 90도)
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

cv2.imshow('Rotated image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 가우시안 블러
blurred_image = cv2.GaussianBlur(image, (31, 31), 0)

cv2.imshow('Blurred image', blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Canny 에지 검출 (임계값 하한성/상한선)
edges = cv2.Canny(image, threshold1=100, threshold2=200)

cv2.imshow('Edges', edges)  # 에지(흰색) / 배경(검정색)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이진화 처리 (임계값 127 기준 흑백으로 변환)
ret, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

cv2.imshow('Binary Image', binary_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 얼굴 탐지기 로드
cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'  # OpenCV 기본 제공 얼굴 캐스케이드 경로
face_cascade = cv2.CascadeClassifier(cascade_path)  # Haar cascade 분류기 로드
```

```python
# 이미지 읽기 및 그레이스케일 변환
image = cv2.imread('face.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 얼굴 검출을 위해 흑백 변환

# 얼굴 탐지
faces = face_cascade.detectMultiScale(
    gray_image,       # 입력 이미지(흑백)
    scaleFactor=1.1,  # 스케일 피라미트 축소 비율 (작으면 촘촘/느림, 크면 빠름/미검출 가능성)
    minNeighbors=5    # 최종 얼굴로 인정할 최소 이웃 후보 수
)

# 탐지된 얼굴에 바운딩 박스 그리기
for (x, y, w, h) in faces:
    cv2.rectangle(
        image,
        (x, y),           # 좌상단 좌표
        (x + w, y + h),   # 우하단 좌표
        (0, 255, 0),      # 초록색 (RGB)
        2  # 선 두께
    )
    

# 결과 이미지 표시
cv2.imshow('Face Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# HOG 디스크립터 초기화
hog = cv2.HOGDescriptor()  # HOG 특징 추출기 생성
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())  # 기본 보행자 검출용 SVM 가중치 로드

# 이미지 읽기
image = cv2.imread('pedestrians.jpg')

# 보행자 검출 (멀티스케일로 보행자 후보 박스 탐지)
(rects, weights) = hog.detectMultiScale(
    image,
    winStride=(4, 4),  # 슬라이딩 윈도우 이동 간격 (작으면 촘촘/느림)
    padding=(8, 8),    # 윈도우 주변 패딩
    scale=1.5  # 이미지 피라미드 스케일 비율 (작으면 촘촘/느림)
)

# 검출된 바운딩 박스 그리기
for (x, y, w, h) in rects:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)  # 초록박스

# 결과 이미지 표시
cv2.namedWindow('Pedestrian Detection', cv2.WINDOW_NORMAL)  # 창 크기 조절 가능하도록 생성
cv2.resizeWindow('Pedestrian Detection', 1000, 700)  # 창 크기를 1000x700으로 설정
cv2.imshow('Pedestrian Detection', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 동영상 파일에서 영상 받아오기
cap = cv2.VideoCapture('video.mp4')

# 웹캠에서 영상 받아오기
# cap = cv2.VideoCapture(0)
```

```python
while cap.isOpened():  # 비디오 캡쳐 객체가 열려있는 동안 반복
    ret, frame = cap.read()  # 다음 프레임 읽기 (성공여부, 이미지 반환)
    if not ret:
        break

    # 프레임 처리 (예: 그레이스케일 변환)
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # 흑백으로 변환

    # 결과 프레임 표시
    cv2.imshow('Video', gray_frame)  # 처리된 프레임을 'Video' 창에 출력

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 자원 해제
cap.release()  # 비디오 캡쳐 장치/파일 해제
cv2.destroyAllWindows()  # 열린 모든 창 닫기
cv2.waitKey(1)  # 창이 닫히지 않는 경우 사용
```

```python
# 웹캠 실시간 YOLO 객체탐지 (OpenvCV)
import cv2  # OpenCV
import numpy as np
from ultralytics import YOLO

model = YOLO('yolo11n.pt')  # 사전학습 YOLO

# 클래스 라벨별 색상 설정 함수
def get_colors(num_colors):  # 클래스 개수만큼 랜덤 색상 팔레트 생성
    np.random.seed(0)
    return [tuple(np.random.randint(0, 255, 3).tolist()) for _ in range(num_colors)]  # (B, G, R) 형태 색상 리스트

class_names = model.names  # 클래스 id -> 클래스명 dict
num_classes = len(class_names)
colors = get_colors(num_classes)
print(class_names)

# 웹캠 프레임(ndarray)에 탐지 결과를 그려서 반환하는 함수
def detect_objects(image: np.array):
    """
    카메라에서 프레임이 전송될 때 이미 ndarray이므로 PIL Image로 변환하지 않는다.
    """
    results = model(image, verbose=False)  # 프레임 1장에 대해 객체탐지 수행
    class_names = model.names

    for result in results:
        boxes = result.boxes.xyxy  # 박스 좌표 (x1, y1, x2, y2)
        confs = result.boxes.conf  # 신뢰도(confidence)
        cls = result.boxes.cls     # 클래스 id
        for box, confidence, class_id in zip(boxes, confs, cls):
            x1, y1, x2, y2 = map(int, box)  # 좌표를 int형으로 변환
            label = class_names[int(class_id)]  # 클래스id -> 클래스명
            cv2.rectangle(image, (x1, y1), (x2, y2), colors[int(class_id)], 2)
            cv2.putText(
                image,
                f'{label} {confidence:.2f}',
                (x1, y1),
                cv2.FONT_HERSHEY_SIMPLEX,  # 폰트 지정
                0.9, (255, 0, 0), 2        # 크기, 색상, 두께
            )
    return image

# 카메라에서 프레임 단위로 캡처
cap = cv2.VideoCapture(0)  # 기본 웹캠(0번) 열기
if not cap.isOpened():
    print('카메라를 찾을 수 없습니다.')

while cap.isOpened():
    ret, frame = cap.read()  # 프레임 1장 읽음 (ret=False면 실패)
    if not ret:
        break

    result_image = detect_objects(frame)  # 프레임에 객체탐지 결과 박스/라벨을 오버레이

    # 프레임을 화면에 표시
    cv2.imshow('Frame', np.array(result_image))

    if cv2.waitKey(1) & 0xFF == ord('q'):  # 1ms 키 입력 대기 후 q면 종료 (입력값은 정수 사용)
        break

# 리소스 해제
cap.release()  # 카메라 장치 해제
cv2.destroyAllWindows()  # OpenCV 창 닫기
cv2.waitKey(1)  # 창이 닫히지 않는 경우 사용
```

```python
from tensorflow.keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# мҰқк°• м—Ҷмқҙ лӘЁлҚё н•ҷмҠө
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense  # CNN кё°ліё л Ҳмқҙм–ҙл“Ө

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),  # 3x3 н•„н„° 32к°ң нҠ№м§•л§ө м¶”м¶ң(мһ…л Ҙ 32, 32, 3)
    MaxPooling2D((2, 2)),  # H, W нҒ¬кё°лҘј м Ҳл°ҳмңјлЎң мӨ„мһ„
    Flatten(),  # (H, W, C) -> (H*W*C) 1м°Ёмӣҗ лІЎн„°лЎң нҺјм№Ё
    Dense(64, activation='relu'),  # мқҖлӢүмёө 64
    Dense(10, activation='softmax')  # 10к°ң нҒҙлһҳмҠӨ нҷ•лҘ 
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',  # м •мҲҳ лқјлІЁ кё°л°ҳ лӢӨмӨ‘л¶„лҘҳ мҶҗмӢӨн•ЁмҲҳ
    metrics = ['accuracy']
)

# н”Ҫм…Җ м •к·ңнҷ” н•ҷмҠө, кІҖмҰқлҚ°мқҙн„°лҠ” 10%
model.fit(x_train / 255.0, y_train, epochs=10, validation_split=0.1)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale = 1./255,           # н”Ҫм…Җ к°’ мҠӨмјҖмқјл§Ғ
    rotation_range = 15,        # -15 ~ +15 лһңлҚӨ нҡҢм „
    width_shift_range = 0.1,    # к°ҖлЎңл°©н–Ҙ мөңлҢҖ 10% лһңлҚӨ мқҙлҸҷ
    height_shift_range = 0.1,   # м„ёлЎңл°©н–Ҙ мөңлҢҖ 10% лһңлҚӨ мқҙлҸҷ
    zoom_range = 0.1,           # -10 ~ 10% лһңлҚӨ нҷ•лҢҖ
    horizontal_flip = True      # мўҢмҡ° л°ҳм „ лһңлҚӨ
)

model.fit(
    datagen.flow(x_train, y_train, batch_size = 64),  # лҚ°мқҙн„° мҰқк°• + м •к·ңнҷ”лҗң л°°м№ҳ мҠӨнҠёлҰј мғқм„ұ
    epochs = 10,
    validation_data = (x_test / 255.0, y_test)  # кІҖмҰқ лҚ°мқҙн„°лҠ” мҰқк°• x
)
```

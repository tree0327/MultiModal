# Vision Transformer(ViT)

Vision Transformer(ViT)는 기존에 자연어 처리(NLP) 분야에서 주로 사용되던 Transformer 구조를 이미지 처리(Computer Vision) 분야에 적용한 딥러닝 모델이다. 2020년 Google Research에서 처음 발표되었으며, 이미지 분류 등 다양한 비전 작업에서 뛰어난 성능을 보이고 있다.

Vision Transformer는 이미지를 패치로 나누어, 각 패치를 토큰으로 처리해 Transformer 구조로 이미지를 이해하는 모델이다. 이로써 이미지 내 전역적 관계를 효과적으로 학습할 수 있다.

- 이미지를 **작은 패치(patch)** 단위로 분할하여 각각을 토큰(token)처럼 취급한다.
- 각 패치는 1차원 벡터로 변환(Flatten)된 후, 임베딩(Embedding) 과정을 거쳐 Transformer의 입력으로 사용된다.
- 각 패치 임베딩에 **위치 정보(Positional Encoding)** 를 더해 순서 정보를 보완한다.
- 이 입력들을 Transformer Encoder에 통과시켜 이미지 전체의 특징을 추출한다.
- 마지막에 **분류 토큰(CLS token)** 의 출력을 사용하여 최종 예측(분류 등)을 수행한다.

**구조**

| 단계             | 설명                                                        |
|------------------|-------------------------------------------------------------|
| 이미지 패치 분할   | 이미지를 일정 크기의 패치로 나눔                              |
| 패치 임베딩      | 각 패치를 1차원 벡터로 변환 후 임베딩                         |
| 위치 임베딩      | 각 패치에 위치 정보를 더함                                    |
| Transformer 인코더 | 여러 Self-Attention 레이어를 거치며 이미지 전체의 관계 학습      |
| 분류 토큰        | 전체 이미지를 대표하는 토큰을 통해 최종 분류 결과 도출          |

**주요 특징**

- **CNN과의 차이점**: CNN은 합성곱 연산을 통해 지역적 특징을 추출하는 반면, ViT는 Self-Attention을 통해 이미지 전체 패치 간의 관계를 한 번에 파악한다.
- **데이터 요구량**: ViT는 CNN보다 사전학습(pre-training) 시 더 많은 데이터가 필요하다. 대규모 데이터셋에서 강력한 성능을 보인다.
- **전이 학습(Transfer Learning)**: 대용량 데이터셋에서 사전학습 후, 소규모 데이터셋에 전이학습하면 좋은 성능을 얻을 수 있다.

## google/vit-base-patch16-224

`google/vit-base-patch16-224` 모델은 구글이 2020년 "An Image is Worth 16x16 Words" 논문에서 발표한 **Vision Transformer (ViT)** 아키텍처의 표준 모델이다. 이 모델은 기존의 CNN(합성곱 신경망) 대신 자연어 처리(NLP)에서 주로 쓰이던 Transformer 구조를 이미지 처리용으로 변형하여 적용한 것이 특징이다.


* **작동 방식**: 이미지를 가로세로 **16x16 픽셀 크기의 패치(Patch)**로 나눈 뒤, 각 패치를 텍스트의 '단어(Token)'처럼 취급하여 Transformer Encoder에 입력한다.
* **사전 학습**: ImageNet-21k(약 1,400만 장, 21,843개 클래스) 데이터셋으로 대규모 사전 학습을 거친 후, ImageNet-1k 데이터셋으로 미세 조정(Fine-tuning)되었다.
* **장점**: 데이터가 방대할수록 CNN보다 성능이 우수하며, 이미지 전체의 전역적 정보(Global Context)를 파악하는 능력이 뛰어나다.

**구글 ViT 모델 라인업 비교**

구글은 모델의 크기(Base, Large, Huge)와 패치 크기, 해상도에 따라 다양한 변체를 제공한다.

**_모델 규모별 비교 (Base vs Large vs Huge):_**

| 모델명 | Hidden Size | Layers | Heads | Parameters | 특징 |
| --- | --- | --- | --- | --- | --- |
| **ViT-Base** | 768 | 12 | 12 | ~86M | 표준 성능과 효율성의 균형 |
| **ViT-Large** | 1024 | 24 | 16 | ~307M | 높은 정확도 요구 시 사용 |
| **ViT-Huge** | 1280 | 32 | 16 | ~632M | 초대형 데이터셋 학습용 |

**_패치 크기(Patch Size):_**

모델명 뒤의 `patch16` 또는 `patch32`는 이미지를 나누는 기본 단위를 의미한다.

* **Patch 16 ($16 \times 16$)**: 이미지를 더 세밀하게 쪼갠다. 토큰 수가 많아져 연산량은 늘어나지만, 정확도가 상대적으로 높다.
* **Patch 32 ($32 \times 32$)**: 이미지를 크게 쪼갠다. 연산 효율성이 좋아 속도가 빠르지만, 세부 정보를 놓칠 수 있다.

**_해상도(Resolution):_**

모델명 끝의 숫자는 입력 이미지의 해상도를 의미한다.

* **224**: 기본 해상도 ( 픽셀). 일반적인 분류 작업에 최적화되어 있다.
* **384**: 고해상도 ( 픽셀). 미세한 특징을 잡아내야 할 때 사용하며, 224 모델을 기반으로 파인튜닝된 경우가 많다.


> 분류 결과의 라벨이 여러 이름인 이유는 동일한 동물의 여러 이름(동의어)을 모두 라벨로 관리하기 때문이다.
>
> ImageNet 같은 대형 이미지 데이터셋에서는 하나의 클래스에 여러 동의어(별칭, 학명등)를 쉼표(,)로 구분해 묶어둔다.

```python
# Huggingface ViT 이미지 분류 모델 / 프로세스 로드
from transformers import ViTImageProcessor, ViTForImageClassification  # ViT 전처리기 / 분류모델 클래스
from PIL import Image  # 이미지 로딩 및 처리(Pillow)
import requests        # URL로 이미지 다운로드

url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
image = Image.open(requests.get(url, stream=True).raw)  # URL에서 이미지 스트림을 받아 PIL 이미지로 열기

processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')  # 사전학습된 ViT 전처리기 로드(리사이즈 / 정규화 등)
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')  # 사전학습된 ViT 분류모델
```

```python
model.config.id2label  # 분류 모델의 클래스 인덱스(id): 문자열 매핑한 dict
```

```python
# ViT로 이미지 분류 추론 및 예측 라벨 출력
inputs = processor(images=image, return_tensors="pt")  # PIL 이미지를 ViT 입력 텐서로 전처리

outputs = model(**inputs)  # 전처리된 입력으로 순전파(추론) 진행
logits = outputs.logits    # 각 클래스 (1000개)에 대한 로짓값 텐서
print(logits.shape)        # (Batch 1, 클래스 1000개)

predicted_class_idx = logits.argmax(-1).item()  # 마지막 차원 기준 최대값 인덱스를 선택
print("Predicted class:", model.config.id2label[predicted_class_idx])  # 라벨 문자열 반환
```

```python
# Transformers pipeline으로 ViT 이미지 분류기 생성
import torch
from transformers import pipeline

# ViT 이미지 분류 파이프라인
classifier = pipeline(
    task = 'image-classification',          # 작업 유형 : 이미지 분류
    model = 'google/vit-base-patch16-224',  # 사전학습된 ViT 모델
    dtype = torch.float16,                  # float16 로드로 VRAM/메모리 사용량 감소
    # device=0,                             # GPU 사용시에 0번 GPU 사용
    device_map = 'auto'                     # accelerate가 장치를 자동(여러 장치/CPU/GPU)으로 배치 (device와 동시 사용 불가)
)
```

```python
results = classifier('http://images.cocodataset.org/val2017/000000039769.jpg')
display(results)  # 예측 결과 리스트(라벨/점수)를 화면에 표시

print(f"{results[0]['label']} ({results[0]['score']})")  # 1등 예측한 라벨과 점수 출력
```
